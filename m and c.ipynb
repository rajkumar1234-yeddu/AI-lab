```python
class Mcb():
    
    def validate(self,x):
        a,b=x
        if(a[0]>=a[1] and b[0]>=b[1]):
            return True
        elif((a[0]==0) or (b[0]==0)):
            return True
        return False
    def successors(self,t,close,x):
        if(self.validate(x)):
            if(x not in close):
                t.append(x)
                close.append(x)
    def rules(self,p1,p2,Open,close):
        t=[]
        #rule 1 2m
        if(p1[2]==1):#rule for boat on left
            if(p1[0]>1):
                x=((p1[0]-2,p1[1],0),(p2[0]+2,p2[1],1))
                self.successors(t,close,x)
        else:#rule for boat on right
            if(p2[0]>1):
                x=((p1[0]+2,p1[1],1),(p2[0]-2,p2[1],0))
                self.successors(t,close,x)       
                            
        #rule 2 1m,1c
        if(p1[2]==1):#rule for boat on left
            if(p1[0]>0 and p1[1]>0):
                x=((p1[0]-1,p1[1]-1,0),(p2[0]+1,p2[1]+1,1))
                self.successors(t,close,x)
        else:#rule for boat on right
                if(p2[0]>0 and p2[1]>0):
                    x=((p1[0]+1,p1[1]+1,1),(p2[0]-1,p2[1]-1,0))
                    self.successors(t,close,x)
                            
        #rule 3 2c
        if(p1[2]==1):#rule for boat on left
            if(p1[1]>1):
                x=((p1[0],p1[1]-2,0),(p2[0],p2[1]+2,1))
                self.successors(t,close,x)
        else:#rule for boat on right
            if(p2[1]>1):
                x=((p1[0],p1[1]+2,1),(p2[0],p2[1]-2,0))
                self.successors(t,close,x)
                            
        #rule 4 1m
        if(p1[2]==1):#rule for boat on left
            if(p1[0]>0):
                x=((p1[0]-1,p1[1],0),(p2[0]+1,p2[1],1))
                self.successors(t,close,x)
        else:#rule for boat on right
            if(p2[0]>0):
                x=((p1[0]+1,p1[1],1),(p2[0]-1,p2[1],0))
                self.successors(t,close,x)
            
        #rule 5 1c
        if(p1[2]==1):#rule for boat on left
            if(p1[1]>0):
                x=((p1[0],p1[1]-1,0),(p2[0],p2[1]+1,1))
                self.successors(t,close,x)
        else:#rule for boat on right
            if(p2[1]>0):
                x=((p1[0],p1[1]+1,1),(p2[0],p2[1]-1,0))
                self.successors(t,close,x)
        return t
            
    def bfs(self,start,goal):
        Open,close=[start],[start]
        p=start
        while(p!=goal and Open):
            p=Open.pop(0)
            p1,p2=p
            t=self.rules(p1,p2,Open,close)
            Open.extend(t)
            print(p)
            print()
    def dfs(self,start,goal):
        Open,close=[start],[start]
        p=start
        while(p!=goal and Open):
            p=Open.pop()
            p1,p2=p
            t=self.rules(p1,p2,Open,close)
            Open.extend(t[::-1])
            print(p)
            print()
                            
x=int(input("enter m c"))
start=((x,x,1),(0,0,0))
goal=((0,0,0),(x,x,1))
g=Mcb()
print("BFS")
g.bfs(start,goal)
print("DFS")
g.dfs(start,goal)
```
